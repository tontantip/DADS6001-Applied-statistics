```r
age<-c(4,3,2,4,3,5,5,4,2,3)
mcost<-c(148,145,128,143,133,159,154,142,118,127)
data_table <- data.frame(age,mcost)
data_table
```

```r
plot(data_table)
```

```r
model <- lm(mcost ~ age, data = data_table)
summary(model)
```

predict age = 3.5

```r
new.age<-data.frame(age= 3.5)
predict(model, newdata = new.age)
```

  mcost  X = 3.5  

```r
predict(model, newdata = new.age, interval = "confidence")
```

  mcost  X = 3.5  :

```r
predict(model, newdata = new.age, interval = "prediction")
```
