# Diagnosis and Plaque Status on Brain Lithium (Two-way anova)

## prepare data frame

```r
nci_np <- c(0.851, 1.554, 1.162, 1.501, 0.840, 0.054, 1.081)
nci_p  <- c(1.965, 2.162, 2.203, 2.641, 2.260, 2.233, 2.185)
ad_np  <- c(0.990, 1.010, 1.874, 1.026, 0.126)
ad_p   <- c(3.428, 3.720, 3.743, 4.491, 2.246)


two_way_data <- data.frame(
  Value = c(nci_np, nci_p, ad_np, ad_p),
  Diagnosis = factor(c(rep("NCI", length(nci_np) + length(nci_p)),
                       rep("AD", length(ad_np) + length(ad_p)))),
  Plaque_Type = factor(c(rep("NP", length(nci_np)),
                         rep("P", length(nci_p)),
                         rep("NP", length(ad_np)),
                         rep("P", length(ad_p))))
)
```

## Test assumption

### Normality test

```r
library(rstatix)

two_way_data %>%
  group_by(Diagnosis, Plaque_Type) %>%
  shapiro_test(Value)
```

### Homogeneity test

```r
res_2way <- aov(Value ~ Diagnosis * Plaque_Type, data = two_way_data)

levene_test(res_2way)
```

## Statistic test

```r
summary(res_2way)
```

## Post hoc (Tukey test)

```r
TukeyHSD(res_2way)
```
